## Read The Data

```python
import pandas as pd
df = pd.read_excel(r"Nasr City Restaurants.xlsx")
```

## View and Drop Resturants In the Same Coordinates With the Same Title

```python
df[df.duplicated(subset=['title', 'location/lat', 'location/lng'], keep=False)].sort_values(by=['location/lat', 'location/lng'])
```

```python
df = df.sort_values(by='reviewsCount', ascending=False).drop_duplicates(subset=['title', 'location/lat', 'location/lng'], keep='first')
```

```python
words = df['title'].str.split(expand=True).stack().value_counts()

words[words.index.str.contains('Kosh|kush|كشري', case=False)]
```

```python
words = df['cuisine'].str.split(expand=True).stack().value_counts()

words[words.index.str.contains('Kosh|kush|كشري', case=False)]
```

## Flag all Koshari Restaurants

```python
df['is_koshari'] = df['title'].str.contains('Kosh|kush|كشري', case=False)
```

```python
df_koshari = df[df['is_koshari']].reset_index(drop=True)
```

```python
df_koshari
```

## Finding the Closest Pair of Koshary Restaurants using the Euclidean Distance Between Two Points
$$  d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2} $$

```python
import numpy as np

min_distance = float('inf') 
closest_pair = None

def euclidean_distance(lat1, lng1, lat2, lng2):
    return np.sqrt((lat2 - lat1)**2 + (lng2 - lng1)**2)

# Iterate over each Koshary restaurant
for i, row1 in df_koshari.iterrows():
    for j, row2 in df_koshari.iterrows():
        if i != j: 
            distance = euclidean_distance(row1['location/lat'], row1['location/lng'], 
                                           row2['location/lat'], row2['location/lng'])
            if distance < min_distance:
                min_distance = distance
                closest_pair = (i, j)

if closest_pair:
    closest_restaurants = df_koshari.loc[[closest_pair[0], closest_pair[1]]]

closest_restaurants
```

```python
print("The two closest Koshary restaurants are:")
print(f"1. {closest_restaurants.iloc[0]['title']} at ({closest_restaurants.iloc[0]['location/lat']}, {closest_restaurants.iloc[0]['location/lng']})")
print(f"2. {closest_restaurants.iloc[1]['title']} at ({closest_restaurants.iloc[1]['location/lat']}, {closest_restaurants.iloc[1]['location/lng']})")
```
